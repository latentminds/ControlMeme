```python
import cv2
```

```python
text_to_add = "Meme.Koll.ai"
path_image = '../images/newmeme2/1bgs.jpg'
path_output = '../images/watermarked.jpg'

image = cv2.imread(path_image)

# add text to image on the bottom right
font = cv2.FONT_HERSHEY_SIMPLEX
bottomLeftCornerOfText = (10, 10)
fontScale = 1
fontColor = (255, 255, 255)
lineType = 2
image_with_text = cv2.putText(image, text_to_add, bottomLeftCornerOfText, font, fontScale, fontColor, lineType)
cv2.imwrite(path_output, image_with_text)
```
